```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

### Base de dados:
* Fonte: https://charts.coinmetrics.io/network-data/

```python
data = pd.read_excel("doge.xlsx")
data.head()
```

```python
df = data.rename(columns={'Time': 'data', 'DOGE / USD Denominated Closing Price': 'valor'})
```

```python
recente = df.tail(23)
recente_data = recente.data
recente_valor = recente.valor
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 6))
plt.title('Doge')
plt.plot(recente["data"], recente['valor'])
plt.xlabel('Periodo', fontsize=18)
plt.ylabel('Valor em USD($)', fontsize=18)
plt.show()
```
