### Base de dados Online:

A escolha por utilizar este tipo de base é o fato de sempre pegar dados atualizados diariamente.

```python
df = web.DataReader('DOGE-USD', data_source='yahoo', start='2021-05-01', end='2021-06-01')
df.head()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 6))
plt.title('Doge')
plt.plot(df['Close'])
plt.xlabel('Periodo', fontsize=18)
plt.ylabel('Valor em USD($)', fontsize=18)
plt.show()
```

>regressão linear
<br> usando a documentação https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
rang = np.arange(0,1,.03125)
reg = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=None).fit(df,rang)
```

```python
reg.score(df, rang)
```

```python
reg.coef_
```

```python
reg.intercept_
```
