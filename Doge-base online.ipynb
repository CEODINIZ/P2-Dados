### Base de dados Online:

A escolha por utilizar este tipo de base Ã© o fato de sempre pegar dados atualizados diariamente.

```python
# df = web.DataReader('BTC-USD', data_source='yahoo', start='2017-07-14', end='2020-05-09')
df = web.DataReader('DOGE-USD', data_source='yahoo', start='2021-05-01', end='2021-05-24')
df.head()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 6))
plt.title('Doge')
plt.plot(df['Close'])
plt.xlabel('Periodo', fontsize=18)
plt.ylabel('Valor em USD($)', fontsize=18)
plt.show()
```
