### Base de dados: O início do escopo

A escolha por utilizar este tipo de base é o fato de sempre pegar dados atualizados diariamente. No entanto, podemos admitir que Doge nem sempre foi uma cryptomoeda tida como relevante, pois originalmente era apenas um meme. 

Para mostrar isso, vamos identificar qual é o mês onde o preço começa a flutuar de maneira significativa para que possamos treinar melhor nosso modelo.

#### Dezembro/2020

```python
df = web.DataReader('DOGE-USD', data_source='yahoo', start='2020-12-01', end='2020-12-31')
df.head()
```

```python
#df = df.reset_index(drop=False)
#df.head()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 6))
plt.title('Doge em Dezembro 2020')
plt.plot(df['Close'])
plt.xlabel('Periodo', fontsize=18)
plt.ylabel('Valor em USD($)', fontsize=18)
plt.show()
```

#### Técnica Aplicada: Regressão Linear
<br> Usando https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

Usamos regressão linear pra tentar prever a cotação da DogeCoin para o próximo mês, com base nas cotações atuais e passadas. Podemos admitir que, devido à instabilidade do assunto do estudo, é possível que esta previsão nao seja muito precisa, pois o mercado de cripto-ativos, é fortemente influenciado por comentários públicos de membros relevantes da comunidade, como por exemplo tweets do Elon Musk em relacao ao Bitcoin e especialmente Dogecoin

```python
rang = np.arange(0,1,0.03125) #array de escala, 31 dias
reg = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=None).fit(df,rang) #regressao linear
```

```python
#score(X, y[, sample_weight])
#Return the coefficient of determination R square of the prediction.
reg.score(df, rang)
```

##### Predição do Doge com base nos dados do mês anterior

```python
predicao = reg.predict(df) #retorna os valores previstos
pred = list(predicao)

#pred2 = pred
```

```python
#lista_pred2 = []
#for e in pred2:
#    lista_pred2.append(e/10)
```

```python
tam = np.arange(0, len(pred))

plt.style.use('fivethirtyeight')
plt.figure(figsize=(12, 6))
plt.title('Previsão do Doge em Janeiro 2021')
plt.plot(tam, pred)
plt.xlabel('Periodo (dias)', fontsize=18)
plt.ylabel('Valor em USD($)', fontsize=18)
plt.show()
```
